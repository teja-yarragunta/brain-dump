```python
# CLASS & OBJECT
# Classes are blueprints (abstract concept of something like blue print)
# Objects are the actual things you built (make something out of the blueprints)
# movie can be a class and the actual movie 'Oppenheimer' is an object
# variables inside the class => attributes
# functions inside the class => methods
# for classes use Pascal Casing
```

```python
# the first thing to give is to give it some variable which are then called attributes, and we give it those attributes when we create the function when it's born
# we do that with the initialization statement, which is the action that creates an object (def __init__)
# init as in initialization
class Movie:
    def __init__(self, title, year, imdb_score, have_seen):
        self.title = title
        self.year = year
        self.imdb_score = imdb_score
        self.have_seen = have_seen
        # the keyword self refers to the object that we are creating, so in the class when we create this new object, that is the self
        # we can call it anything but convention is to use self
    def nice_print(self):
        print("Title: ", self.title)
        print("Year of production: ", self.year)
        print("IMDB Score: ", self.imdb_score)
        print("I have seen it: ", self.have_seen)
```

```python
# creating the object
film_1 = Movie("Oppenheimer", 2023, 9, True)
film_2 = Movie("Barbie", 2023, 8, True)
print(film_1.title, film_2.imdb_score)
# call using the method of the class
film_2.nice_print() # this doesn't take an argument that's because it's run on film_2, so the argument is implicit in that call, because it's done from that object, which is equivalent to the below
Movie.nice_print(film_2)
```

```python
films = [film_1,film_2]
print(films[1].title,films[0].title)
films[0].nice_print()
```
