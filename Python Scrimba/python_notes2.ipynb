```python
# - Create a distance converter converting Km to miles
# - Take two inputs from user: Their first name and the distance in km
# - Print: Greet user by name and show km, and mile values
# - 1 mile is 1.609 kilometers
# - hint: use correct types for calculating and print
# - Did you capitalize the name
name = input('Enter your name: ')
distance_km = input('Enter distance in km: ')
distance_mi = float(distance_km)/1.609
print(f'Hi {name.title()}! {distance_km}km is equivalent to {round(distance_mi,1)} miles.')
```

```python
# slicing
s = "Python"
# Positive indexing
# P  y  t  h  o  n
# 0  1  2  3  4  5

# Negative indexing
# P   y   t   h   o   n
#-6  -5  -4  -3  -2  -1

# python strings, lists are zero indexed
msg="hello i'm John."
print(msg[0])
print(msg[1])
print(msg[2])
print(msg[13])
print(msg[-0]) # same as msg[0]
print(msg[-1])
print(msg[-2])
print(msg[-3])
print(msg[-4])
```

```python
print(msg[2:]) # from 2nd index(inclusive) to every thing - zero index based
print(msg[2::])
print(msg[2:4]) # from 2nd index to 4th index (exclusive)
```

```python
msg = 'hello'
print(msg[:4])
print(msg[:-1])
print(msg[::-1])
```

```python
s = "Python"
print(s[::2])   # Pto  (every 2nd char)
print(s[1::2])  # yhn
print(s[::-1])  # nohtyP (reverse string)
```
