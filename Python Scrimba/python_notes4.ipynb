```python
# tuples - faster Lists you can't change
friends = ['John','Michael','Terry','Eric','Graham']
friends_tuple = ('John','Michael','Terry','Eric','Graham')
print(friends)
print(friends_tuple)
print(friends[2])
print(friends_tuple[2])
print(friends[2:4])
print(friends_tuple[2:4])
# tuples are immutable
# they are also slightly faster to work with when iterating, searches and all
# pop, remove, append functions don't work
friends_tuple.append('teja')

```

```python
# Sets - blazingly fast unordered Lists - no duplicates
friends = ['John','Michael','Terry','Eric','Graham']
friends_tuple = ('John','Michael','Terry','Eric','Graham')
friends_set = {'John','Michael','Terry','Eric','Graham','Eric', 'Eric'}
print(friends)
print(friends_tuple)
print(friends_set)

# sets are much faster than lists
```

```python
# intersection in sets
friends_set = {'John','Michael','Terry','Eric','Graham','Eric'}
my_friends_set = {'Reg','Loretta','Colin','Eric','Graham'}
print(friends_set.intersection(my_friends_set))
```

```python
# difference in sets
print(friends_set.difference(my_friends_set))
```

```python
# union in sets
print(friends_set.union(my_friends_set))
```
